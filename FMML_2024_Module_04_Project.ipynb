## Gradient Descent

````python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
````

````python
## TASK-8 : Code the same things as they were done for the first two examples
import matplotlib.pyplot as plt
import numpy as np

# Define the function and its derivative
def g(x):
    return x**2  # Example: g(x) = x^2

def dg(x):
    return 2*x  # Example: derivative of g(x) = x^2

# Define the gradient descent function
def gradient_descent(derivative_func, initial_guess, multiplier=0.01, precision=0.0001, max_iter=500):
    new_x = initial_guess
    previous_x = 0
    x_list = [new_x]
    deriv_list = [derivative_func(new_x)]
    cost_list = [g(new_x)]

    for _ in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        x_list.append(new_x)
        deriv_list.append(gradient)
        cost_list.append(g(new_x))

        if abs(new_x - previous_x) < precision:
            break

    return new_x, x_list, deriv_list, cost_list

# Values for x_2
x_2 = np.linspace(-3, 3, 400)

# Run gradient descent with different learning rates
low_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.0005, precision=0.0001, max_iter=500)
mid_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.001, precision=0.0001, max_iter=500)
high_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.002, precision=0.0001, max_iter=500)

# Experiment
insane_gamma = gradient_descent(derivative_func=dg, initial_guess=1.9, multiplier=0.25, precision=0.0001, max_iter=500)

# Plot function and reduction in cost for each iteration
plt.figure(figsize=[20, 15])

# Plot low gamma
plt.subplot(3, 2, 1)
plt.xlim(-3, 3)
plt.ylim(0, 10)
plt.title('Low Gamma (0.0005)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.plot(x_2, g(x_2), color='blue', linewidth=3, alpha=0.8)
plt.scatter(low_gamma[1], g(np.array(low_gamma[1])), color='red', s=100, alpha=0.6)

plt.subplot(3, 2, 2)
plt.title('Reduction in cost (Low Gamma)', fontsize=17)
plt.xlabel('Iteration', fontsize=16)
plt.ylabel('Cost', fontsize=16)
plt.plot(low_gamma[3], color='green', linewidth=3, alpha=0.8)

# Plot mid gamma
plt.subplot(3, 2, 3)
plt.xlim(-3, 3)
plt.ylim(0, 10)
plt.title('Mid Gamma (0.001)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.plot(x_2, g(x_2), color='blue', linewidth=3, alpha=0.8)
plt.scatter(mid_gamma[1], g(np.array(mid_gamma[1])), color='red', s=100, alpha=0.6)

plt.subplot(3, 2, 4)
plt.title('Reduction in cost (Mid Gamma)', fontsize=17)
plt.xlabel('Iteration', fontsize=16)
plt.ylabel('Cost', fontsize=16)
plt.plot(mid_gamma[3], color='green', linewidth=3, alpha=0.8)

# Plot high gamma
plt.subplot(3, 2, 5)
plt.xlim(-3, 3)
plt.ylim(0, 10)
plt.title('High Gamma (0.002)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.plot(x_2, g(x_2), color='blue', linewidth=3, alpha=0.8)
plt.scatter(high_gamma[1], g(np.array(high_gamma[1])), color='red', s=100, alpha=0.6)

plt.subplot(3, 2, 6)
plt.title('Reduction in cost (High Gamma)', fontsize=17)
plt.xlabel('Iteration', fontsize=16)
plt.ylabel('Cost', fontsize=16)
plt.plot(high_gamma[3], color='green', linewidth=3, alpha=0.8)

plt.tight_layout()
plt.show()

# Print results for each learning rate
print("Low Gamma (0.0005):", low_gamma[0])
print("Mid Gamma (0.001):", mid_gamma[0])
print("High Gamma (0.002):", high_gamma[0])
print("Insane Gamma (0.25):", insane_gamma[0])

````

#### Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````
